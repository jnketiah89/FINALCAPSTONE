# Which districts have the largest number of students? Top 10

```python
filtered_df = merged_absenteeism[(merged_absenteeism['Year'] >= 2018) & (merged_absenteeism['Year'] <= 2022)]
```

```python
district_totals = filtered_df.groupby('system_name')['n_students'].sum().reset_index()
```

```python
sorted_districts = district_totals.sort_values('n_students', ascending=False)
```

```python
top_10_districts = sorted_districts.head(10)
```

```python
print(top_10_districts)
```

# Racial demographics for these top 10 districts

```python
filtered_df = merged_absenteeism[(merged_absenteeism['Year'] >= 2018) & (merged_absenteeism['Year'] <= 2022) & (merged_absenteeism['system_name'].isin(top_10_districts['system_name']))]
```

```python
district_race_totals = filtered_df.groupby(['system_name', 'subgroup'])['n_students'].sum().reset_index()
```

```python
print(district_race_totals)
```

# Perecntage of each subgroup represented in the top 10 districts

```python
district_totals = district_race_totals.groupby('system_name')['n_students'].sum()
```

```python
district_race_totals['percentage'] = (district_race_totals['n_students'] / district_race_totals['system_name'].map(district_totals)) * 100
```

```python
print(district_race_totals)
```

```python
davidson_df = merged_absenteeism[(merged_absenteeism['system_name'] == 'Davidson County') & (merged_absenteeism['subgroup'].isin(['Black or African American', 'White', 'Hispanic', 'Asian']))]
```

```python
race_totals = davidson_df.groupby('subgroup')['n_students'].sum()
```

```python
race_percentages = (race_totals / race_totals.sum()) * 100
```

```python
plt.figure(figsize=(8, 8))
plt.pie(race_percentages, labels=race_percentages.index, autopct='%1.1f%%', startangle=10)
plt.title('Subgroups in Davidson County')
plt.axis('equal')
plt.show()
```
